```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
import tensorflow as tf
from tensorflow import keras as K
```

```python
from cnn import cnn_predict
import detect
model_path = "models/cnn_model.h5"
```

### 导入图片

```python
image_path = "docs/images/plate.jpg"
image_src = cv2.imread(image_path, cv2.IMREAD_COLOR)
detect.imshow(image_src)
```

## 图片处理

```python
mask, ori_gray = detect.get_mask(image_src)
plt.subplot(1,2,1)
detect.imshow(ori_gray)
plt.subplot(1,2,2)
detect.imshow(mask)
```

```python
lic_img = detect.locate_and_correct(image_src, mask)
detect.imshow(lic_img[0])
```

## 字符识别

```python
model = K.models.load_model(model_path)
```

```python
lic_pred = cnn_predict(model, lic_img)
```

```python
detect.imshow(lic_pred[0][0])
print(lic_pred[0][-1])
```
